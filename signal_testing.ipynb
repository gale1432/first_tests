```python
def create_model():
    model = keras.models.Sequential([
        keras.Input(shape=(22,5,7)),
        keras.layers.Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
        keras.layers.MaxPooling1D(pool_size=2),
        keras.layers.Conv1D(filters=64, kernel_size=2, activation='relu', padding='same'),
        keras.layers.Conv1D(filters=64, kernel_size=1, activation='relu', padding='same'),
        keras.layers.Flatten(),
        keras.layers.LSTM(10),
        keras.layers.LSTM(10),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.Dense(12, activation='softmax')
    ])
    return model

def create_model_2D():
    model = keras.models.Sequential([
        keras.Input(shape=(22,5,7)),
        keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu', padding='same'),
        keras.layers.MaxPooling2D(pool_size=2),
        keras.layers.Conv2D(filters=64, kernel_size=2, activation='relu', padding='same'),
        keras.layers.Conv2D(filters=64, kernel_size=1, activation='relu', padding='same'),
        keras.layers.Reshape((11, 128)),
        keras.layers.LSTM(10),
        #keras.layers.LSTM(10),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.Dense(12, activation='softmax')
    ])
    return model
```

```python
model = create_model_2D()
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, epochs=5, verbose=1)
```
